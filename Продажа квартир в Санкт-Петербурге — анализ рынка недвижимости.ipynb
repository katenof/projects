## Предобработка данных

Начнем работать с пропусками. Первым юудет столбец "is_apartment".

```python
#посчитаем кол-во пропусков
print(apart['is_apartment'].isna().sum())
apart['is_apartment'].value_counts()
```

```python
apart['is_apartment']=apart['is_apartment'].fillna(0)
apart.head()
```

```python
#теперь проработаем данные с балконами
apart['balcony'] = apart['balcony'].fillna(0)
print(apart['balcony'].isnull().sum())
#теперь данные по балконам не отсуствуют
```

```python
#теперь попробуем заменить пропуски в ceiling_height на медиану значений
#apart['ceiling_height'] = apart['ceiling_height'].fillna(apart['ceiling_height'].median())
apart['ceiling_height'] = statistics.median([apart['ceiling_height']])
apart.head(20)
```

```python
# "is_apartment" у нас со значениями 0 и 1, а "studio" нет. Заменим эти значения для более приятного просомтра данных.
apart['studio'] = apart['studio'].map({True: 1, False: 0})
# сделаем тоже самое с "open_plan"
apart['open_plan'] = apart['open_plan'].map({True: 1, False: 0})
apart.head()
```

После того как мы заменили данные, которые возможно было заменить, стоит привести остальные данные в порядок.

```python
#начнем со столбца "first_day_exposition" и приведем его к новому формату
apart['first_day_exposition'] = pd.to_datetime(apart['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
apart.head()
```

Теперь нужно округлить данные некоторых столбцов. Столбцы 'total_area','living_area', 'kitchen_area' нужно округлить до 1-го знака полсе запятой, а столбцы 'living_area_part', 'kitchen_area_part', 'ceiling_height' до 2-х знаков после запятой.

```python
list_1 = ['total_area', 'living_area', 'kitchen_area']
for column in list_1:
    apart[column] = apart[apart[column].notnull()][column].round(1)
apart.head(10)
```

```python
list_2 = ['living_area', 'kitchen_area', 'ceiling_height']
for column in list_2:
    apart[column] = apart[apart[column].notnull()][column].round(2)
apart.head(10)
```

```python
#Столбец "last_price" приведем к тысячам. 
apart['last_price'] = (apart[apart['last_price'].notnull()]['last_price']/1000*1.0).round().astype(int)
apart.head()
```

"В остальных столбцах отсуствие данных скорее всего означает их буквальное отсутсвие (нет рядом парков и т.д), а значит не имеет смысла трогать эти пропуски" 

Теперь, когда данные проработы, можно перейти к расчетам.

## Расчёты и добавление результатов в таблицу

Теперь нам нужно рассчитать и добавить в таблицу цена квадратного метра; день недели, месяц и год публикации объявления; этаж квартиры; варианты — первый, последний, другой; соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
#теперь нужно добавить еще и день, но перед эти создадим словарь, чтобы изменить день с числа на название дней недели.
day_week = {1:'понедельник', 2:'вторник', 3:'среда', 4:'четверг', 5:'пятница', 6:'суббота', 7:'воскресенье'}
apart['day'] = apart['first_day_exposition'].dt.day_name()
apart.head(10)
```

```python
#чтобы построить нормальную диаграмму зависимости удаленности от центра, 
#нам нужно написать функцию для категоризации значений 'cityCenters_nearest'
def fillna_ceiling_height(row):
    if row['cityCenters_nearest'] < 2500:
        return 'центр'
    elif row['cityCenters_nearest'] < 5000:
        return 'близко к центру'
    elif row['cityCenters_nearest'] < 17000:
        return 'в городе'
    elif row['cityCenters_nearest'] >= 17000:
        return 'область'

apart['fromcenter_cat'] = apart.apply(fillna_ceiling_height, axis=1)
```
